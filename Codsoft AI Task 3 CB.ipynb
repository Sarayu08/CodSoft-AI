```python
import re

def food_delivery_chatbot():
    print("Hi! I am your Food Delivering Chatbot. I can answer simple questions and help you order delicious food. What can I do for you today?")

    while True:
        user_input = input("You: ")

        # Convert user input to lowercase for case-insensitive matching
        user_input_lower = user_input.lower()

        # Check if the user wants to end the conversation
        if re.search(r'\b(goodbye|bye|exit)\b', user_input_lower):
            print("Goodbye! If you have more questions or want to order again, feel free to ask.")
            break

        # Check if the user greets the chatbot
        if re.search(r'\b(hi|hello|hey)\b', user_input_lower):
            print("Hello! I'm your Food Delivering Chatbot. I can assist you with food orders. What type of cuisine are you in the mood for today?")
            continue
        # Check if the user asks about available cuisines
        if re.search(r'\b(what type of cuisine is available|tell some types|what are available)\b', user_input_lower):
            print("We offer a variety of cuisines! Our menu includes Pizza, Burger, Pasta, Sushi, Tacos, Indian dishes, and more. What type of cuisine are you interested in?")
            continue
        # Check if the user expresses interest in ordering other cuisines
        if re.search(r'\b(i would like to order other cuisine|i would like to order more|can you take another order)\b', user_input_lower):
            print("Sure! We have a diverse menu with various cuisines. What specific dish or type of cuisine would you like to order?")
            continue
     
        # Define rules and responses related to different cuisines
        if re.search(r'\b(pizza)\b', user_input_lower):
            print("Great choice! We have Margherita, Pepperoni, Veggie Supreme, and BBQ Chicken pizzas. What size would you like?")
            size_input = input("You: ")
            print(f"Perfect! I'll add a {size_input} {user_input_lower} to your order. May I know your delivery location?")
            delivery_location = input("You: ")
            print(f"Got it! Your order will be delivered to {delivery_location}. Could you please provide your phone number for updates?")
            phone_number = input("You: ")
            print(f"Thank you! Your order will be on its way to {delivery_location}. We will contact you at {phone_number} for any updates.")

        elif re.search(r'\b(burger)\b', user_input_lower):
            print("Awesome! Our burger menu includes Classic, BBQ, Veggie, and Spicy Chicken options. What type of burger do you prefer?")
            burger_type = input("You: ")
            print(f"Great! What size would you like your {burger_type} burger?")
            burger_size = input("You: ")
            print(f"Excellent choice! I'll add a {burger_size} {burger_type} burger to your order. May I know your delivery location?")
            delivery_location = input("You: ")
            print(f"Got it! Your order will be delivered to {delivery_location}. Could you please provide your phone number for updates?")
            phone_number = input("You: ")
            print(f"Thank you! Your order will be on its way to {delivery_location}. We will contact you at {phone_number} for any updates.")

        elif re.search(r'\b(pasta)\b', user_input_lower):
            print("Pasta it is! We offer Alfredo, Bolognese, Pesto, and Carbonara pasta dishes. Which one are you craving?")
            pasta_type = input("You: ")
            print(f"Great! What size would you like your {pasta_type} pasta?")
            pasta_size = input("You: ")
            print(f"Excellent choice! I'll add a {pasta_size} {pasta_type} pasta to your order. May I know your delivery location?")
            delivery_location = input("You: ")
            print(f"Got it! Your order will be delivered to {delivery_location}. Could you please provide your phone number for updates?")
            phone_number = input("You: ")
            print(f"Thank you! Your order will be on its way to {delivery_location}. We will contact you at {phone_number} for any updates.")

        elif re.search(r'\b(sushi)\b', user_input_lower):
            print("Sushi time! We have a variety of rolls, including California Roll, Dragon Roll, Spicy Tuna Roll, and Salmon Nigiri. What would you like?")
            sushi_type = input("You: ")
            print(f"Fantastic! What size would you like your {sushi_type} sushi?")
            sushi_size = input("You: ")
            print(f"Fantastic choice! I'll add a {sushi_size} {sushi_type} to your order. May I know your delivery location?")
            delivery_location = input("You: ")
            print(f"Got it! Your order will be delivered to {delivery_location}. Could you please provide your phone number for updates?")
            phone_number = input("You: ")
            print(f"Thank you! Your order will be on its way to {delivery_location}. We will contact you at {phone_number} for any updates.")

        elif re.search(r'\b(taco)\b', user_input_lower):
            print("Taco time! Our taco options include Chicken, Beef, Vegetarian, and Fish tacos. What type of taco are you in the mood for?")
            taco_type = input("You: ")
            print(f"Great choice! I'll add some {taco_type} tacos to your order. May I know your delivery location?")
            delivery_location = input("You: ")
            print(f"Got it! Your order will be delivered to {delivery_location}. Could you please provide your phone number for updates?")
            phone_number = input("You: ")
            print(f"Thank you! Your order will be on its way to {delivery_location}. We will contact you at {phone_number} for any updates.")

        elif re.search(r'\b(indian)\b', user_input_lower):
            print("Craving Indian cuisine? Our menu includes Butter Chicken, Chicken Biryani, Saag Paneer, and Tandoori Chicken. What dish would you like?")
            indian_dish = input("You: ")
            print(f"Delicious choice! I'll add {indian_dish} to your order. May I know your preferred spice level?")
            spice_level = input("You: ")
            print(f"Great! I'll make sure your {indian_dish} is prepared with a spice level of {spice_level}. May I know your delivery location?")
            delivery_location = input("You: ")
            print(f"Got it! Your order will be delivered to {delivery_location}. Could you please provide your phone number for updates?")
            phone_number = input("You: ")
            print(f"Thank you! Your order will be on its way to {delivery_location}. We will contact you at {phone_number} for any updates.")
       # Ask if the user wants to add beverages
            print("Would you like to add any beverages to your order?")
            add_beverages_input = input("You: ")
            if re.search(r'\b(yes|sure)\b', add_beverages_input.lower()):
                print("Great! We have a variety of beverages. For sodas, we have Coke, Pepsi, Sprite, and Dr. Pepper. What type of soda would you like to add?")
                soda_type = input("You: ")
                print(f"Fantastic choice! What size would you like for your {soda_type}? (Small, Medium, Large)")
                soda_size = input("You: ")
                print(f"Excellent! I'll add a {soda_size} {soda_type} to your order.")
                
                # Ask for juice
                print("For juices, we have Orange, Apple, Grape, Muskmelon, and Pineapple. What type of juice would you like to add?")
                juice_type = input("You: ")
                print(f"Great! What size would you like for your {juice_type}? (Small, Medium, Large)")
                juice_size = input("You: ")
                print(f"Excellent! I'll add a {juice_size} {juice_type} to your order.")
                
                # Ask for shake
                print("If you prefer shakes, we offer Chocolate,Nutella, Vanilla, Strawberry, KitKat, Belgian Chocolate, and Oreo shakes. What type of shake would you like to add?")
                shake_type = input("You: ")
                print(f"Delicious choice! What size would you like for your {shake_type}? (Small, Medium, Large)")
                shake_size = input("You: ")
                print(f"Perfect! I'll add a {shake_size} {shake_type} to your order. Is there anything else I can help you with?")
            else:
                print("Okay, let me know if there's anything else I can assist you with.")



        else:
            print("I'm sorry, I didn't catch that. Could you please specify a cuisine or ask about something else?")

# Start the conversation
food_delivery_chatbot()
```
